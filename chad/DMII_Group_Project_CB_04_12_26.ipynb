```python
import pandas as pd
from pathlib import Path

mimic_path = Path("data/MIMIC/processed")

if not mimic_path.exists():
    print("❌ MIMIC folder not found!")
else:
    cleaned_count = 0
    for entity_dir in sorted(mimic_path.iterdir()):
        if entity_dir.is_dir():
            text_file = entity_dir / "text.csv"
            if text_file.exists():
                df = pd.read_csv(text_file)
                # Find the timestamp column (most common names in MIMIC preprocessing)
                timestamp_col = None
                for possible in ['date_time', 'charttime', 'admittime', 'time', 'timestamp']:
                    if possible in df.columns:
                        timestamp_col = possible
                        break

                if timestamp_col is None:
                    print(f"⚠️  No timestamp column found in {entity_dir.name}")
                    continue

                # Rename to the exact name the library expects
                df_clean = df[[timestamp_col, 'text']].copy()
                df_clean = df_clean.rename(columns={timestamp_col: 'date_time'})

                # Save back
                df_clean.to_csv(text_file, index=False)
                cleaned_count += 1
                print(f"✅ Fixed: {entity_dir.name}  (timestamp: {timestamp_col} → date_time)")
            else:
                print(f"⚠️  No text.csv in {entity_dir.name}")

    print(f"\n🎉 Finished! Cleaned {cleaned_count} MIMIC text.csv files.")
    print("Now each file has exactly two columns: ['date_time', 'text']")
```
